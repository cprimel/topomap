```julia
using Random, Polyhedra, GLMakie
```

```julia
x = rand(Float64, (3,100))
println(typeof(x))
#scatter(x[1,:],x[2,:],x[3,:])
```

```julia
import GLPK

v_x = vrep(transpose(x))
removevredundancy(v_x, GLPK.Optimizer)
```

```julia
p = polyhedron(v_x)
```

```julia
m = Polyhedra.Mesh{3}(p)
```

```julia
GLMakie.mesh(m, color=:blue)
```
